```python
import pandas as pd 
from bs4 import BeautifulSoup
import requests
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers = headers ).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all("div", class_="companyCardWrapper__primaryInformation")
```

```python
len(company)
```

```python
name = []
for i in company:
    name.append((i.find('h2',class_="companyCardWrapper__companyName").text.strip()))
```

```python
ratting = (soup.find_all('span',class_="companyCardWrapper__companyRatingCount"))
ratings = []
for i in ratting:
    ratings.append(i.text.strip())
```

```python
ratings
```

```python
detail = []
details = (soup.find_all('span',class_='companyCardWrapper__interLinking'))
for i in details:
    detail.append(i.text.strip())
```

```python
detail
```

```python
print(name[0] + "\n" + ratings[0] + "\n" + detail[0])
```

```python
d = {
    "name": name,
    "rating": ratings,
    "detail": detail
}
df = pd.DataFrame(d)
```

```python
final = pd.DataFrame()

for page in range(1, 15):

    url = f"https://www.ambitionbox.com/list-of-companies?page={page}"
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    name = []
    ratings = []
    detail = []
    companies = soup.find_all("div", class_="companyCardWrapper__primaryInformation")
    for company in companies:
        name.append(company.find('h2', class_="companyCardWrapper__companyName").text.strip())

    rating_tags = soup.find_all('span', class_="companyCardWrapper__companyRatingCount")
    for r in rating_tags:
        ratings.append(r.get_text(strip=True))

    detail_tags = soup.find_all('span', class_="companyCardWrapper__interLinking")
    for d in detail_tags:
        detail.append(d.get_text(strip=True))
    df = pd.DataFrame({
        "name": name,
        "rating": ratings,
        "detail": detail
    })

   
    final = pd.concat([final, df], ignore_index=True)

final
```
